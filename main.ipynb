### <b><span style='color:#16C2D5'>|</span> Connaissances du domaine</b>

<br>


1. **`Age`:** L'√¢ge est un facteur important pour pr√©dire le risque de diab√®te. Avec l'√¢ge, le risque de d√©velopper un diab√®te augmente. Cela est d√ª en partie √† des facteurs tels qu'une activit√© physique r√©duite, des changements dans les niveaux d'hormones et une plus grande probabilit√© de d√©velopper d'autres conditions de sant√© qui peuvent contribuer au diab√®te.

2. **Le sexe** : Le sexe peut jouer un r√¥le dans le risque de diab√®te, bien que l'effet puisse varier. Par exemple, les femmes ayant des ant√©c√©dents de diab√®te gestationnel (diab√®te pendant la grossesse) ont un risque plus √©lev√© de d√©velopper un diab√®te de type 2 plus tard dans la vie. 3. **L'indice de masse corporelle (IMC):** L'IMC est une mesure de la graisse corporelle bas√©e sur la taille et le poids d'une personne. Il est couramment utilis√© comme indicateur du poids global et peut √™tre utile pour pr√©dire le risque de diab√®te. Un IMC √©lev√© est associ√© √† une plus grande probabilit√© de d√©velopper un diab√®te de type 2. L'exc√®s de graisse corporelle, en particulier autour de la taille, peut entra√Æner une r√©sistance √† l'insuline et entraver la capacit√© de l'organisme √† r√©guler le taux de sucre dans le sang.

4. **Hypertension art√©rielle:** L'hypertension art√©rielle est une affection qui coexiste souvent avec le diab√®te. Ces deux affections pr√©sentent des facteurs de risque communs et peuvent contribuer √† leur d√©veloppement respectif. L'hypertension augmente le risque de d√©velopper un diab√®te de type 2 et vice versa. Les deux affections peuvent avoir des effets n√©fastes sur la sant√© cardiovasculaire.

5. **Les maladies cardiaques, y compris les maladies coronariennes et l'insuffisance cardiaque, sont associ√©es √† un risque accru de diab√®te. La relation entre les maladies cardiaques et le diab√®te est bidirectionnelle, ce qui signifie que le fait d'√™tre atteint d'une maladie augmente le risque de d√©velopper l'autre. Cela s'explique par le fait qu'elles partagent de nombreux facteurs de risque communs, tels que l'ob√©sit√©, l'hypertension art√©rielle et l'hypercholest√©rol√©mie.

6. **"Ant√©c√©dents de tabagisme "** Le tabagisme est un facteur de risque modifiable pour le diab√®te. On a constat√© que la cigarette augmentait le risque de d√©velopper un diab√®te de type 2. Arr√™ter de fumer peut r√©duire de mani√®re significative le risque de d√©velopper un diab√®te et ses complications.

7. **Le taux d'HbA1c:** L'HbA1c (h√©moglobine glyqu√©e) est une mesure de la glyc√©mie moyenne au cours des 2 ou 3 derniers mois. Elle fournit des informations sur le contr√¥le √† long terme de la glyc√©mie. Des taux d'HbA1c plus √©lev√©s indiquent un moins bon contr√¥le de la glyc√©mie et sont associ√©s √† un risque accru de d√©velopper un diab√®te et ses complications.

8. **La glyc√©mie est la quantit√© de glucose (sucre) pr√©sente dans le sang √† un moment donn√©. Une glyc√©mie √©lev√©e, en particulier √† jeun ou apr√®s avoir consomm√© des hydrates de carbone, peut indiquer un trouble de la r√©gulation du glucose et augmenter le risque de d√©velopper un diab√®te. ### <b>I <span style='color:#16C2D5'>|</span> Introduction </b>

Dans cette analyse, nous avons choisi le classifieur RandomForest comme mod√®le. **<marque style="background-color:#16C2D5;color:white;border-radius:5px;opacity:1.0">L'algorithme RandomForest</marque>** est une m√©thode d'apprentissage d'ensemble qui fonctionne en construisant une multitude d'arbres de d√©cision pendant la formation et en produisant la classe qui est le mode des classes pour la classification ou la pr√©diction moyenne des arbres individuels pour la r√©gression.

Plusieurs raisons ont guid√© notre choix de **<span style='color:#16C2D5'>Random Forest</span>** pour cette t√¢che :

1. **Traitement de donn√©es volumineuses:** **<span style='color:#16C2D5'>Random Forest</span>** est capable de traiter efficacement des ensembles de donn√©es volumineux √† haute dimensionnalit√©. Notre jeu de donn√©es, qui contient un nombre important de lignes et plusieurs caract√©ristiques, entre dans cette cat√©gorie.

2. **Robustesse face au surajustement:** **<span style='color:#16C2D5'>Random Forest</span>** r√©duit le risque de surajustement, qui est un probl√®me courant avec les arbres de d√©cision. L'algorithme y parvient en cr√©ant un ensemble de **<marque style="background-color:#DDE6ED;color:black;border-radius:4px;opacity:1.0">arbres de d√©cision</marque>** (une "for√™t") et en effectuant la pr√©diction finale sur la base du vote majoritaire des arbres individuels.

3. **Gestion des types de donn√©es mixtes:** Dans notre ensemble de donn√©es, nous avons √† la fois des caract√©ristiques num√©riques et cat√©gorielles. **<span style='color:#16C2D5'>Random Forest</span>** g√®re ces m√©langes en douceur, ce qui en fait un choix id√©al.

4. **Importance des caract√©ristiques:** **<span style='color:#16C2D5'>Random Forest</span>** offre un moyen simple d'estimer l'importance des caract√©ristiques. √âtant donn√© que notre objectif est d'√©tudier l'impact de diff√©rents facteurs sur le diab√®te, cette caract√©ristique est particuli√®rement utile.

5. **Non-lin√©arit√©:** Les donn√©es m√©dicales contiennent souvent des relations complexes et non lin√©aires. **<span style='color:#16C2D5'>Random Forest</span>**, √©tant un mod√®le non lin√©aire, peut capturer ces relations de mani√®re efficace.

### <b>II <span style='color:#16C2D5'>|</span> Encodage</b>

### ‚ùì Question : Impl√©mentation de l'encodage √† chaud (One-Hot Encoding)

- **√âcrivez le code n√©cessaire pour effectuer un encodage √† chaud sur des variables sp√©cifiques dans le DataFrame `data` :**
  - D√©finissez une fonction `perform_one_hot_encoding` qui prend en param√®tres un DataFrame (`df`) et le nom d'une colonne (`column_name`). La fonction doit :
    - Effectuer un encodage √† chaud sur la colonne sp√©cifi√©e et cr√©er des colonnes indicatrices (dummies).
    - Supprimer la colonne originale du DataFrame et ajouter les nouvelles colonnes indicatrices.
    - Retourner le DataFrame modifi√©.
  - Utilisez cette fonction pour effectuer un encodage √† chaud sur la variable `gender` dans `data`.
  - R√©p√©tez l'op√©ration pour la variable `smoking_history` dans `data`.


```python
# D√©finition de la fonction pour l'encodage √† chaud
def perform_one_hot_encoding(df, column_name):     # column_name correspond au nom de la colonne que l'on va encoder avec pd.get_dummies()
    
    # Cr√©ation des colonnes indicatrices (dummies)
    dummies = pd.get_dummies(df[column_name], prefix = column_name).astype(int)
    # pd.get_dummies() : Cr√©e une colonne pour chaque cat√©gorie unique de la variable (permet de convertir ces textes en nombres 0/1 pour chaque cat√©gorie)
    # prefix=column_name ajoute le nom de la colonne d‚Äôorigine comme pr√©fixe (ex: gender_male, gender_female)
    
    # Suppression de la colonne originale, les nouvelles colonnes cr√©√©es par pd.get_dummies() (gender_Male, gender_Female) restent dans le DataFrame
    df = df.drop(column_name, axis=1)

    # Ajout des nouvelles colonnes indicatrices (sert √† ajouter les nouvelles colonnes cr√©√©es par pd.get_dummies() au DataFrame original.)
    df = pd.concat([df, dummies], axis=1)   # pd.concat([...], axis=1) ‚Üí concat√®ne les colonnes horizontalement

    return df

# Exemple d'utilisation sur les colonnes 'gender' et 'smoking_history'
# La fonction perform_one_hot_encoding retourne un DataFrame modifi√©, mais ne change pas df automatiquement si on n‚Äôutilise pas df = ...
data = perform_one_hot_encoding(df, 'gender')
data = perform_one_hot_encoding(data, 'smoking_history')

# V√©rification du r√©sultat
data.head(10)
    
    
```

```python
# D√©finir le r√©√©chantillonnage pour sur-√©chantillonner (SMOTE) avec un taux de surv√©chantillonnage de 10%
# La classe minoritaire doit repr√©senter 10% de la classe majoritaire apr√®s sur√©chantillonnage ‚áí sampling_strategy = 0.10
over = SMOTE(sampling_strategy = 0.10)

# D√©finir le r√©√©chantillonnage pour sous-√©chantillonner (RandomUnderSampler) avec un taux de sous-√©chantillonnage de 50%
# La classe minoritaire doit repr√©senter 50% de la classe majoritaire apr√®s sous-√©chantillonnage ‚áí sampling_strategy = 0.5
under = RandomUnderSampler(sampling_strategy = 0.50)
```

### <b> Pr√©traitement : Scaler</span></b> 

Le pr√©traitement est une √©tape cruciale avant l'entra√Ænement du mod√®le. Dans ce cas, les caract√©ristiques num√©riques sont normalis√©es (moyenne supprim√©e et mise √† l'√©chelle pour obtenir une variance unitaire), et les caract√©ristiques cat√©gorielles sont cod√©es √† une vitesse. **<span style='color:#16C2D5'>La normalisation</span>** n'est pas n√©cessaire pour tous les mod√®les, mais il s'agit g√©n√©ralement d'une bonne pratique. **<span style='color:#16C2D5'>Le codage √† une vitesse</span>** est n√©cessaire pour que les variables cat√©gorielles soient correctement comprises par le mod√®le d'apprentissage automatique.

Le **<marque style="background-color:#16C2D5;color:white;border-radius:5px;opacity:1.0">StandardScaler</mark>** dans sklearn est bas√© sur l'hypoth√®se que les donn√©es, Y, suivent une distribution qui n'est pas n√©cessairement gaussienne (normale), mais nous la transformons tout de m√™me de mani√®re √† ce que sa distribution ait une valeur moyenne de 0 et un √©cart-type de 1.</p> 

En d'autres termes, √©tant donn√© un vecteur de caract√©ristiques <em>x</em>, il modifie les valeurs comme suit:</p>

<p class="formulaDsp">
\[ Y_i = \frac{x_i - \mu(\vec{x})}{\sigma(\vec{x})} \]
</p>

**o√π:**
<ul>
<li>\( x_i \) est le i-√®me √©l√©ment du vecteur de caract√©ristiques original \( \vec{x} \),</li>
<li>\( \mu(\vec{x}) \) est la moyenne du vecteur d'√©l√©ments, et</li>
<li>\( \sigma(\vec{x}) \) est l'√©cart type du vecteur de caract√©ristiques.</li>
</ul>

</p> <p>Les donn√©es transform√©es \N( Y \N) (chaque \N( Y_i \N)) auront des propri√©t√©s telles que \N( mean(Y) = 0 \N) et \N( std(Y) = 1 \N).

> Cette transformation est √©galement connue sous le nom de normalisation du score Z.


```python
# D√©finir le pr√©processeur
preprocessor = ColumnTransformer(    # ColumnTransformer est une classe de Scikit-Learn qui permet √† la fois des variables num√©riques et cat√©gorielles, car on doit souvent leur appliquer des traitements distincts.
    transformers=[
        ('num', StandardScaler(), ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level','hypertension','heart_disease']),   # valeurs num√©riques -> valeurs continues normalis√©es
        ('cat', OneHotEncoder(), ['gender','smoking_history'])   # valeurs cat√©gorielles -> colonnes binaires (0/1)
    ])

# Diviser les donn√©es en caract√©ristiques et variable cible
X = df.drop('diabetes', axis=1)
y = df['diabetes']
```

```python
# Cr√©er un pipeline qui pr√©traite les donn√©es, r√©√©chantillonne les donn√©es, puis entra√Æne un classifieur
clf = imbPipeline(steps=[('preprocessor', preprocessor),   # Scale les colonnes num√©riques et encode les colonnes cat√©gorielles
                      ('over', over),  # over : Sur-√©chantillonne la classe minoritaire avec SMOTE
                      ('under', under),  # under : Sous-√©chantillonne la classe majoritaire pour √©quilibrer
                      ('classifier', RandomForestClassifier())])  # Entra√Æne un mod√®le Random Forest sur les donn√©es transform√©es
```

### Construction du mod√®le et r√©glage des hyperparam√®tres

Un pipeline est construit qui applique d'abord les √©tapes de pr√©traitement et entra√Æne ensuite un mod√®le sur les donn√©es. Nous utilisons un **<mark style="background-color:#16C2D5;color:white;border-radius:5px;opacity:1.0">RandomForestClassifier</mark>**, qui est un algorithme populaire et puissant pour les t√¢ches de classification. Les hyperparam√®tres du mod√®le sont ajust√©s √† l'aide de **<mark style="background-color:#16C2D5;color:white;border-radius:5px;opacity:1.0">GridSearchCV</mark>** , qui effectue une recherche exhaustive sur les valeurs de param√®tres sp√©cifi√©es pour l'estimateur. Le mod√®le le plus performant est s√©lectionn√© sur la base d'une validation crois√©e.

```python
# Nous d√©finissons par d√©faut ces hyperparam√®tres
param_grid = {
    'classifier__n_estimators': [50, 100, 200],  # D√©finit le nombre d‚Äôarbres dans la for√™t al√©atoire. Ici on teste 3 tailles : 50, 100, 200 arbres.
    'classifier__max_depth': [None, 10, 20],    # Profondeur maximale des arbres. None ‚Üí pas de limite dans le d√©veloppement (chaque feuille devient pure ou contient < min_samples_split).
    'classifier__min_samples_split': [2, 5, 10],  # Nombre minimum d‚Äô√©chantillons requis pour diviser un noeud interne. Si un noeud contient moins de min_samples_split, il ne sera pas divis√©. Augmenter cette valeur ‚Üí arbres plus simples et moins sujets au surapprentissage.
    'classifier__min_samples_leaf': [1, 2, 4]  # Nombre minimum d‚Äô√©chantillons dans une feuille terminale. Emp√™che que des feuilles contiennent tr√®s peu de donn√©es ‚Üí contr√¥le le surapprentissage. Exemple : 1 ‚Üí feuilles avec 1 √©chantillon possible, 4 ‚Üí chaque feuille doit contenir au moins 4 √©chantillons.
}

# Si min_samples_leaf = 1 : une feuille peut contenir seulement un √©chantillon.
# Cela signifie que l‚Äôarbre peut continuer √† se diviser jusqu‚Äô√† atteindre une tr√®s grande profondeur.
# Avantage : l‚Äôarbre peut parfaitement ¬´ s‚Äôadapter ¬ª aux donn√©es d‚Äôentra√Ænement.
# Inconv√©nient : risque tr√®s √©lev√© de surapprentissage (le mod√®le m√©morise les d√©tails et bruit).

# Si min_samples_leaf = 4 : chaque feuille doit contenir au moins 4 √©chantillons.
# Les arbres seront plus courts et plus robustes.
# Le mod√®le g√©n√©ralise mieux sur des donn√©es qu‚Äôil n‚Äôa jamais vues.

# Avec le param_grid qu'on a √©tabli ici :
# n_estimators ‚Üí 3 options
# max_depth ‚Üí 3 options
# min_samples_split ‚Üí 3 options
# min_samples_leaf ‚Üí 3 options
# Nombre total de combinaisons = 3 √ó 3 √ó 3 √ó 3 = 81
# Chaque combinaison est √©valu√©e via cross-validation pour choisir la meilleure.
```

```python
# Cette cellule met du temps √† s'ex√©cuter, entre 5 et 10 minutes. 
# GridSearchCV permet de :
# Tester toutes les combinaisons possibles des hyperparam√®tres d√©finis dans param_grid.
# Utiliser une validation crois√©e pour √©valuer chaque combinaison de mani√®re robuste.
# S√©lectionner la combinaison qui donne la meilleure performance selon le score choisi (par d√©faut : accuracy).

# Create Grid Search object
grid_search = GridSearchCV(clf, param_grid , cv=2)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)   # Toujours utiliser stratify=y pour des t√¢ches de classification d√©s√©quilibr√©es.
# stratify = y assure que la proportion des classes dans le jeu d‚Äôentra√Ænement et le jeu de test reste la m√™me que dans le dataset original.

# Train the model
grid_search.fit(X_train, y_train)

# Print the best parameters
print("Best Parameters: ", grid_search.best_params_)

# classifier__n_estimators	100 arbres dans la for√™t
# classifier__max_depth	10 (profondeur maximale des arbres)
# classifier__min_samples_split	5 (un noeud doit avoir au moins 5 √©chantillons pour √™tre divis√©)
# classifier__min_samples_leaf	2 (une feuille doit contenir au moins 2 √©chantillons)
```

### <b><span style='color:#16C2D5'>|</span> Interpr√©tez les r√©sultats </b>

Le r√©sultat montre les meilleurs param√®tres pour notre **<span style='color:#16C2D5'>mod√®le de for√™t al√©atoire</span>** qui ont √©t√© trouv√©s pendant le processus de r√©glage des hyperparam√®tres :

1. **max_depth of 10:** Ceci indique que la profondeur maximale des arbres dans la for√™t est de 10 niveaux. Le fait de limiter la profondeur de l'arbre permet de r√©duire l'overfitting. Il ressort de ce r√©sultat qu'un arbre de complexit√© moyenne est le mieux adapt√© √† nos donn√©es. Une trop grande complexit√© (un arbre plus profond) peut capturer le bruit, et une trop faible complexit√© (un arbre moins profond) peut ne pas capturer la structure sous-jacente des donn√©es.

2. **min_√©chantillons_feuille de 2:** Cela signifie que chaque feuille (le n≈ìud final d'un arbre de d√©cision, o√π les pr√©dictions sont faites) doit contenir au moins deux √©chantillons. Ce param√®tre, tout comme max_depth, est un moyen de contr√¥ler le surajustement. En exigeant au moins deux √©chantillons pour effectuer une pr√©diction, le mod√®le √©vite de s'adapter aux valeurs aberrantes ou au bruit dans les donn√©es d'apprentissage.

3. **min_samples_split of 2:** Ceci nous indique qu'un noeud doit contenir au moins deux √©chantillons pour √™tre divis√© (pour cr√©er deux noeuds enfants). Tout comme le param√®tre min_samples_leaf, ce param√®tre peut aider √† contr√¥ler le surajustement.

4. **n_estimateurs de 50:** Il s'agit du nombre d'arbres de d√©cision dans la for√™t. L'algorithme Random Forest fonctionne en faisant la moyenne des pr√©dictions de plusieurs arbres de d√©cision pour obtenir une pr√©diction finale, ce qui permet de r√©duire l'overfitting et la variance. Dans ce cas, il semble que la pr√©sence de 50 arbres dans la for√™t donne les meilleures performances.

<b><span style='color:black'> 4. Matrice de confusion</span></b> </div>

Le mod√®le form√© est √©valu√© sur l'ensemble de test**. La **<marque style="background-color:#16C2D5;color:white;border-radius:5px;opacity:1.0">matrice de confusion</marque>** est utilis√©e pour visualiser les performances du mod√®le. Elle montre les pr√©dictions vraies positives, vraies n√©gatives, fausses positives et fausses n√©gatives du mod√®le.

<br>

![](https://miro.medium.com/v2/resize:fit:1218/1*jMs1RmSwnYgR9CsBw-z1dw.png)

#### **<span style='color:#16C2D5'>Pr√©cision:</span>**

La pr√©cision est une mesure du nombre de pr√©dictions vraies positives qui sont effectivement correctes. Elle est d√©finie comme le nombre de vrais positifs (TP) divis√© par la somme des vrais positifs (TP) et des faux positifs (FP).

<p class="formulaDsp">
\[ Pr√©cision = \frac{TP}{TP + FP} \]
</p>

#### **<span style='color:#16C2D5'>Rappel:</span>**


Le rappel (ou sensibilit√©) est une mesure du nombre de cas positifs r√©els correctement identifi√©s. Il est d√©fini comme le nombre de vrais positifs (TP) divis√© par la somme des vrais positifs (TP) et des faux n√©gatifs (FN).

<br>

<p class="formulaDsp">
\[ Rappel = \frac{TP}{TP + FN} \]
</p>

#### **<span style='color:#16C2D5'>F1-Score:</span>**


Le score F1 est la moyenne harmonique de la pr√©cision et du rappel et tente de trouver un √©quilibre entre la pr√©cision et le rappel. Il est d√©fini comme 2 fois le produit de la pr√©cision et du rappel divis√© par la somme de la pr√©cision et du rappel.

<br>

<p class="formulaDsp">
\[ Score F1 = \frac{2 * Pr√©cision * Rappel}{Pr√©cision + Rappel} \]


<br>
    
<div style="border-radius:10px;border:#16C2D5 solid;padding : 15px;background-color:#ffffff00;font-size:100%;text-align:left">
    <b><span style='color:#16C2D5'>|</span> Dans toutes ces formules : </b>
    
Les vrais positifs (TP) sont les cas dans lesquels nous avons pr√©dit oui (pr√©sence de diab√®te) et o√π le r√©sultat r√©el √©tait √©galement oui.
Les vrais n√©gatifs (TN) sont les cas dans lesquels nous avons pr√©dit non et o√π la r√©alit√© √©tait √©galement non.
Les faux positifs (FP) sont les cas o√π nous avons pr√©dit oui, mais o√π la r√©alit√© √©tait non.
Les faux n√©gatifs (FN) sont les cas o√π nous avons pr√©dit un non, mais o√π le r√©sultat a √©t√© un oui.   
</div>


```python
# Question: Comment pr√©dire les valeurs cibles sur un ensemble de test en utilisant le mod√®le le mieux ajust√© ?
y_pred = grid_search.predict(X_test)    # grid_search.predict(X_test) ‚Üí utilise le meilleur mod√®le trouv√© par GridSearchCV pour pr√©dire les labels du test set.

# Question: Comment afficher la pr√©cision du mod√®le ?
print("Model Accuracy: ", accuracy_score(y_test, y_pred))  # accuracy_score(y_test, y_pred) ‚Üí calcule la pr√©cision globale.

# Question: Comment g√©n√©rer et afficher un rapport d√©taill√© sur les performances du mod√®le ?
print(classification_report(y_test, y_pred))   # classification_report(y_test, y_pred) ‚Üí donne pr√©cision, rappel, F1-score pour chaque classe.

# Question: Quelle fonction utiliser pour calculer la matrice de confusion ?
cm = confusion_matrix(y_test, y_pred)

# Configuration et affichage de la matrice de confusion
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')  # sns.heatmap() ‚Üí visualise la matrice avec des valeurs annot√©es.

# Question: Comment d√©finir le titre de la figure ?
plt.title('Matrice de confusion du mod√®le Random Forest pour la pr√©diction du diab√®te')

# Question: Comment nommer l'axe des abscisses ?
plt.xlabel('Classe pr√©dite')

# Question: Comment nommer l'axe des ordonn√©es ?
plt.ylabel('Classe r√©elle')

# Question: Quelle fonction utiliser pour afficher la figure ?
plt.show()
```

<b><span style='color:black'> 5. Importance de la fonctionnalit√©</span></b> </div>

Enfin, l'importance de chaque caract√©ristique est calcul√©e. Il s'agit de la diminution totale de l'impuret√© du n≈ìud (pond√©r√©e par la probabilit√© d'atteindre ce n≈ìud, qui est approxim√©e par la proportion d'√©chantillons atteignant ce n≈ìud) moyenn√©e sur tous les arbres de l'ensemble. **L'importance des caract√©ristiques donne un aper√ßu des caract√©ristiques les plus utiles pour faire des pr√©dictions.** Les caract√©ristiques sont class√©es en fonction de leur importance et visualis√©es √† l'aide d'un diagramme √† barres.

```python
# Question: Comment obtenir les noms des caract√©ristiques apr√®s un traitement OneHotEncoder dans un pipeline ?
# grid_search.best_estimator_ : C‚Äôest le meilleur mod√®le trouv√© par GridSearchCV, le pipeline complet entra√Æn√© avec les hyperparam√®tres optimaux.
# Ce pipeline contient le pr√©processeur, le sur/sous-√©chantillonnage, et le classificateur.
onehot_columns = list(grid_search.best_estimator_.named_steps['preprocessor']    # named_steps['preprocessor'] donne l‚Äô√©tape de pr√©traitement (le ColumnTransformer qui scale et encode).
                      .named_transformers_['cat']       # named_transformers_['cat'] r√©cup√®re le OneHotEncoder qui s‚Äôapplique aux colonnes cat√©gorielles.
                      .get_feature_names_out(['gender','smoking_history']))  # get_feature_names_out(['gender','smoking_history']) : retourne les noms des nouvelles colonnes qui nt √©t√© encod√©es

# Question: Compl√©tez 
feature_names = ['age', 'BMI', 'HbA1c_level', 'blood_glucose_level', 'hypertension', 'heart_disease'] + list(onehot_columns)
# onehot_columns est g√©n√©ralement un numpy.ndarray apr√®s get_feature_names_out().
# list(onehot_columns) le transforme en liste Python classique, ce qui est plus pratique pour le concat√©ner avec les colonnes num√©riques ou cr√©er un DataFrame

# Question: Comment obtenir les importances des caract√©ristiques d'un classificateur ?
importances = grid_search.best_estimator_.named_steps['classifier'].feature_importances_
# feature_importances_ est un attribut des arbres de d√©cision et for√™ts al√©atoires (comme RandomForestClassifier) dans scikit-learn
# C‚Äôest un tableau de nombres indiquant l‚Äôimportance relative de chaque feature utilis√©e par le mod√®le pour faire ses pr√©dictions.
# Plus la valeur est grande, plus la feature contribue √† la d√©cision du mod√®le.
# Calcul : Chaque arbre dans la for√™t calcule la diminution de l‚Äôimpuret√© (Gini ou Entropie) √† chaque split pour chaque feature.
# Donc l‚Äôimportance d‚Äôune feature est la moyenne de ces diminutions sur tous les arbres. Les valeurs sont ensuite normalis√©es pour que la somme soit √©gale √† 1.

# Question: Comment cr√©er un DataFrame √† partir de deux listes ?
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
# feature_names ‚Üí liste de tous les noms de colonnes que le mod√®le utilise (num√©riques + colonnes cr√©√©es par OneHotEncoder).
# importances ‚Üí valeurs correspondantes de feature_importances_ pour chaque colonne.

# Question: Comment trier un DataFrame par une colonne ?
importance_df = importance_df.sort_values('Importance', ascending = False)
# ascending=False, les features les plus importantes apparaissent en premier, ce qui est tr√®s pratique pour lire rapidement quelles variables influencent le plus la pr√©diction

# Question: Comment afficher les importances des caract√©ristiques ?
print(importance_df)

# Configuration de la figure pour le trac√© des importances
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data= importance_df)
# data comme source de donn√©es pour le graphique

# Question: Comment d√©finir le titre de la figure ?
plt.title('Importance des caract√©ristiques pour la pr√©diction du diab√®te')
# Ce que montre le barplot : le graphique montre quelle variable contribue le plus au mod√®le (parfait aussi pour les publics non-techniques)

# Question: Quelle fonction utiliser pour afficher la figure ?
plt.show()
```
